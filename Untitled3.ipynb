## Add the dataset and preparation the datas for the plot

```python
# add the booking data set
df = pd.read_csv("hotel_radison.csv")
```

```python
# Zeitstempel anpassen (Datenformat)

df['checkin'] = (df['checkin']/1000).apply(datetime.datetime.fromtimestamp)
df['checkout'] = (df['checkout']/1000).apply(datetime.datetime.fromtimestamp)
```

```python
df['checkin']= pd.to_datetime(df.checkin, format='%Y-%m-%d')
df['checkin'] = df.checkin.dt.date

df['checkout']= pd.to_datetime(df.checkout, format='%Y-%m-%d')
df['checkout'] = df.checkout.dt.date

df['timestamp']= pd.to_datetime(df['timestamp']).dt.date
df = df.assign(duration=(df.checkout-df.checkin).dt.days)

df.date = pd.to_datetime(df.date)
df.checkin = pd.to_datetime(df.checkin)
df = df.assign(daysBeforeDeparture=(df.checkin-df.date).dt.days)
```

```python
#Reisedatum und Zeitpunkt der Buchung formatieren
df['holidayMonth'] = df.checkin.dt.month.values
df['calendarWeek'] = df.checkin.dt.week.values 
df['holidayYear'] = df.checkin.dt.year.values
df['holidayMonthname'] = df.checkin.dt.strftime("%B")

df['VACheckMonthName'] = df.date.dt.strftime("%B")
df['VACheckYear'] = df.date.dt.year.values
df['VACheckMonth'] = df.date.dt.month.values
```

```python
# Limit the data of the same start
test = df[(df.date >= '2018-11-25')]
```

```python
# Reisedatum zusammenfügen
test["Trip-Time"] = df["holidayMonthname"].map(str) + ' ' + df["holidayYear"].map(str)
```

```python
# Buchungsdatum/Vacancy Check Datum zusammenfügen
test["VA-Date"] = df["VACheckMonthName"].map(str) + ' ' + df["VACheckYear"].map(str)
```

```python
#Dummy data set erstellen mir den Buchungs/VA-Check und Reisezeiten
dummy =test.groupby(['VA-Date','Trip-Time'])[['VA-Date']].count().rename(columns = {'VA-Date':'counts'}).reset_index()
```

## Christoph's chord example: Vacancy-check data 


```python
data = hv.Dataset((list(df_pivot.columns), list(df_pivot.index), df_pivot),
                  ['target', 'source'], 'value').dframe()
```

```python
data['target'] = data['target'].apply(lambda x: 'Trip-' + x)
data['source'] = data['source'].apply(lambda x: 'VA-' + x)
```

```python
data
```

```python
chord = hv.Chord(data)
chord.opts(
    node_color='index', edge_color='source', label_index='index', 
    cmap='Category20', edge_cmap='Category20', width=500, height=500)
```
